```python
import numpy as np
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt
```

```python
pandasDF = pd.read_csv( "./../data/gig_leases.csv" )
```

```python
T = pandasDF["duration"]
E = pandasDF["censored"]
plt.hist(T, bins = 50)
plt.show()
```

```python
kmf = KaplanMeierFitter()
kmf.fit(durations = T, event_observed = E, label='Move out probablity')
plt.title('GIG Moveout Prediction')
kmf.plot_survival_function()
```

```python
from lifelines import WeibullFitter
wbf = WeibullFitter()
wbf.fit(T, E, label='GIG Survival Graph')
wbf.plot()
```
